### CUSTOMISATION

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([515, 205, 100, 240])
y2 = np.array([486,624,826,846])
y3 = np.array([134,658,246,568])
plt.plot(x, y1, marker=".", markersize=20, markerfacecolor="#55FF00",markeredgecolor="#000000",linestyle="dotted",linewidth='2',color="#000000")
plt.plot(x, y2, marker=".", markersize=20, markerfacecolor="#FF0000",markeredgecolor="#000000",linestyle="dotted",linewidth='2',color="#000000")
plt.plot(x, y3, marker=".", markersize=20, markerfacecolor="#0015FF",markeredgecolor="#000000",linestyle="dotted",linewidth='2',color="#000000")
plt.show()
```

### LABLES

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2023, 2024, 2025, 2026])
y = np.array([515, 205, 100, 240])

plt.title("Learning Trend", fontsize=20, 
          family="serif", 
          fontweight="bold", 
          color="#5174EA")

plt.xlabel('Year', fontsize=12)
plt.ylabel('No. of Students', fontsize=12)

plt.plot(x, y, marker='o', 
         markersize=10, 
         markerfacecolor='#1cd3fc', 
         linewidth=2)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()          # ‚Üê this line is actually optional with %matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])

plt.tick_params(axis="both")
plt.plot(x,y)
plt.xticks(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])

plt.tick_params(axis="both")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.plot(x,y)
plt.xticks(x)
plt.show()
```

### bar chaRts

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])

plt.barh(x,y,color="blue") ## horizontal bar
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])

plt.bar(x,y,color=['red','yellow','blue','green']) ## vertical bar
```

### PI CHART

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])
plt.pie(y, labels=x,autopct='%1.1f%%',startangle=90) 
# autopct	Adds percentage labels 
# colors	Sets colors for pie slices 	plt.pie(values, colors=['red','yellow','blue','green'])
# shadow	Adds drop shadow 
# startangle	Rotates pie chart 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2023,2024,2025,2026])
y=np.array([515,205,100,240])
plt.scatter(x, y,alpha=1,s=200,label='class trennds')   #alpha: transparency, s: marker size
plt.legend()
```

```python
## figure emtire canvas
## ax = a sinngle plot(subplot)
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6,7])
figure,axes = plt.subplots(2,2)
axes[0,0].plot(x,x*2.1,c="red")
axes[0,0].set_title("demo1")
axes[0,1].plot(x,x**2,c="green")
axes[0,1].set_title("demo2")
axes[1,0].bar(x,x*2,color="black")
axes[1,0].set_title("demo3")
axes[1,1].plot(x,x**3,c="blue")
axes[1,1].set_title("demo4")
```

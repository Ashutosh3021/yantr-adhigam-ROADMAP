```python
# yyi = input("enter the year to download data for: ")
# mmi = input("enter the month to download data for: ")
# ddi = input("enter the day to download data for: ")

# yyf = input("enter the year to end data download: ")
# mmf = input("enter the month to end data download: ")
# ddf = input("enter the day to end data download: ")

# Download data
aapl_data = yf.download("AAPL", start=f"2016-01-01", end=f"2026-02-01")
msft_data = yf.download("msft", start=f"2016-01-01", end=f"2026-02-01")
google_data = yf.download("goog", start=f"2016-01-01", end=f"2026-02-01")
NVDA_data = yf.download("NVDA", start=f"2016-01-01", end=f"2026-02-01")
AMZN_data = yf.download("AMZN", start=f"2016-01-01", end=f"2026-02-01")
META_data = yf.download("META", start=f"2016-01-01", end=f"2026-02-01")

BTC_data = yf.download("btc-usd",start=f"2016-01-01", end=f"2026-02-01")
ETH_data = yf.download("eth",start=f"2016-01-01", end=f"2026-02-01")

GLD_data = yf.download("GLD",start=f"2016-01-01", end=f"2026-02-01")
SLV_data = yf.download("SLV",start=f"2016-01-01", end=f"2026-02-01")

# View the first few rows
print(aapl_data.tail())
```

```python
# concatinate the dataframes with propper high,low, open, close, volume and adj close columns with the stock name as prefix
data = pd.concat([aapl_data.add_prefix('AAPL_'), msft_data.add_prefix('MSFT_'), google_data.add_prefix('GOOG_'), NVDA_data.add_prefix('NVDA_'), AMZN_data.add_prefix('AMZN_'), META_data.add_prefix('META_'), BTC_data.add_prefix('BTC_'), ETH_data.add_prefix('ETH_'), GLD_data.add_prefix('GLD_'), SLV_data.add_prefix('SLV_')], axis=1)
data.to_csv('stock_data.csv')

```

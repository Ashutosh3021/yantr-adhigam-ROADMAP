```python
import requests
import pandas as pd
import time

df_list = []

session = requests.Session()

for i in range(1, 500):
    try:
        url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=e290b59391ef9741f5ab9a1064dd733a&language=en-US&page={i}"
        response = session.get(url, timeout=10)
        response.raise_for_status()
        temp_df = pd.DataFrame(response.json()['results'])[
            ["id","title","release_date","overview","popularity","vote_average","vote_count"]
        ]
        df_list.append(temp_df)
        time.sleep(0.4)   # Respect rate limit
    except Exception as e:
        print(f"Error on page {i}: {e}")
        time.sleep(5)
df = pd.concat(df_list, ignore_index=True)

df.head()
```

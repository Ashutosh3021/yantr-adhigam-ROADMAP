```python
## 0 d tensors
import numpy as np
a=np.array(4) # 1 d vector
print(a)
print(a.ndim) # prints the dimentions of the tensor
```

```python
## 1 d tensors
a=np.array([12,23,34,45,56]) # 5 d vector
print(a)
print(a.ndim)
```

```python
## 2 d tensors
a=np.array([[1,2],[3,4]]) ## 4 d vector
print(a)
print(a.ndim)
```
